```python
from qiskit import QuantumCircuit, Aer, assemble
from qiskit.visualization import array_to_latex
from math import pi
import numpy as np
from qiskit.visualization import plot_histogram, plot_bloch_multivector
```

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.cx(0,1)
qc.draw()
```

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.cx(0,1)
display(qc.draw())  # `display` is a command for Jupyter notebooks
                    # similar to `print`, but for rich content

# Let's see the result
svsim = Aer.get_backend('aer_simulator')
qc.save_statevector()
qobj = assemble(qc)
final_state = svsim.run(qobj).result().get_statevector()
plot_bloch_multivector(final_state)
```

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.x(1)
qc.h(1)
display(qc.draw())
# See the result
qc1 = qc.copy()
qc1.save_statevector()
final_state = svsim.run(qc1).result().get_statevector()
display(qiskit.visualization.array_to_latex(final_state, prefix="\\text{Statevector} = "))
plot_bloch_multivector(final_state)
```

```python
qc = QuantumCircuit(2)
qc.cp(pi/4, 0, 1)
qc.draw()
```

```python
usim = Aer.get_backend('aer_simulator')
qobj = assemble(qc)
```

```python
qc = QuantumCircuit(2)
qc.cp(pi/4, 0, 1)
display(qc.draw())
# See Results:
qc.save_unitary()
qobj = assemble(qc)
unitary = usim.run(qobj).result().get_unitary()
array_to_latex(unitary, prefix="\\text{Controlled-T} = \n")
```
